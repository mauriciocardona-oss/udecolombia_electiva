```python
# Importar librerias necesarias para Machine Learning

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV en un Dataframe de Pandas

df = pd.read_csv("datos.csv")
df.head()
```

```python
df.columns
```

```python
# Graficar el scatter plot de los datos

plt.scatter(df["area"], df["precio"], color="blue")
plt.title("Precios Vs Area")
plt.xlabel("Area")
plt.ylabel("Precio")
plt.show()
```

### Predecir el precio de una vivienda que tiene un area 120 m2

```python
# Cargamos la libreria de regresion lineal

from sklearn.linear_model import LinearRegression

# Creamos el modelo (instancia o contener) de regresion lineal
modelo = LinearRegression()
modelo
```

```python
# Exploracion de las dimensiones de los datos
print("El tipo de dato de area es ", type(df['area']))
print("El tipo de dato de area es ", type(df['precio']))
```

```python
print(type(df[['area']]))
```

```python
print(df['area'].shape) # Devuelve un estructura de 2 dimensiones (filas, columnas) 
print(df[['area']].shape) # Devuelve un estructura de 2 dimensiones (filas, columnas)  
print(df['precio'].shape) # Devuelve un estructura de 1 dimension (filas, columnas)
```

```python
# entrenamos el modelo con los datos de area y precio
modelo.fit(df[["area"]], df["precio"])
```

```python
#  puedo hacer la prediccion del precio de una casa con un area de 120 metros cuadrados
area_nueva = 120
precio_predicho = modelo.predict([[area_nueva]])
precio_predicho_modelo = precio_predicho[0]
print(precio_predicho_modelo)
```

Ahora quiero encontrar el mismo valor pero con las variables asociadas a la ecuacion de la linea recta

```python
# Quiero conocer el intercepto del modelo
b = modelo.intercept_
print(b)
```

```python
# Ahora quiero identificar el coeficiente de la variable area

a = modelo.coef_[0]
print(a)
```

```python
# Ahora quiero escribir la ecuacion de la recta que se ajusta a los datos
print("La ecuacion de la recta es: y = ", a, "x + ", b)
```

```python
# Ahora quiero predecir el precio de una casa con area de 120 metros cuadrados
area_nueva = 120
precio_predicho_formula = a * area_nueva + b
```

```python
print("El precio predicho por la formula es: ", precio_predicho_formula)
print("El precio predicho por el modelo es: ", precio_predicho_modelo)
```
